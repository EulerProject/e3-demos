```e3
load tap butterflies.txt
```

```e3
print tap
```

```e3
add articulation 1997.Amauris_hyalites_subsp_hyalites is_included_in 1995.Amauris_damocles_subsp_hyalites
```

e3 associates a <b>state</b> and <b>workspace</b> with a directory in which it is executed. The state is stored in the directory <a target="_blank"href="./.e3">./.e3</a> and the workspace at <a target="_blank" href="./e3_data">./e3_data</a>.

The workspace stores the <ul><li><a target="_blank" href="./e3_data/index.html">html-renderable history</a></li><li>taxonomy alignment problems (e.g. <a target="_blank" href="./e3_data/demo_butterflies">demo_butterflies</a>)</li><li>commands (e.g. <a target="_blank" href="./e3_data/demo_butterflies/graph_worlds">graph worlds</a>)</li><li>execution-time (see history)</li><li>run-time (see history)</li><li>configuration (e.g. <a target="_blank" href="./e3_data/demo_butterflies/graph_worlds/config.txt">config.txt</a>)</li><li>output data (e.g. <a target="_blank" href="./e3_data/demo_butterflies/graph_worlds/graph_worlds_0_mnpw.svg">graph_worlds_0_mnpw.svg</a>)</li><ul>


You can access all of this directly from the history. Take a look at the history by <a target="_blank" href="http://htmlpreview.github.io/?https://github.com/EulerProject/e3-demos/blob/master/demos/history_and_file_system_footprint/e3_data/index.html">rendering it</a> in your browser. You will see that we stepped through a number of taxonomy alignment problems by refining the original problem. Additionally output was produced by executing reasoning commands on some of the problems. You can double click taxonomy alignment problems to analyse the problem, and you can also double click output files to look at them.

The state stores the <ul><li><a target="_blank" href="./.e3/config">configuration</a></li><li><a target="_blank" href="./.e3/style">style</a></li><li><a target="_blank" href="./.e3/current_tap">active taxonomy alignment problem</a></li><li><a target="_blank" href="./.e3/names">taxonomy alignment problem names</a></li><li><a target="_blank" href="./.e3/taps">cache</a> of computed reasoning results, which allows to avoid re-computation</li><li>json-formatted <a target="_blank" href="./.e3/history">history</a></li></ul>

For additional information about the structure of the workspace and  state, see <a target="_blank" href="https://github.com/EulerProject/e3">the e3 project</a>.
