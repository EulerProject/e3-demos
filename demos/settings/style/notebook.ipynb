Here we demonstrate how you can view the style settings, set specific style values and reset all the style values to their default value.

The configuration is stored in YAML format at <a href="./.e3/style">./.e3/style</a> and can also be edited with a text editor.

For more information about the style settings see the <a href="https://github.com/EulerProject/EulerX/wiki/Stylesheet-Processor">Wiki</a> of the EulerX project.


```e3
print style
```

```e3
use tap demo_fossilweevils
```

```e3
graph tap
```

```e3
set style input/graphstyle/legend = off
```

```e3
print style
```

```e3
graph tap
```

```e3
reset style
```

```e3
print style
```
